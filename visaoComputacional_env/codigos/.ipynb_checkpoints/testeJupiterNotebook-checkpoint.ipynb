```python
import cv2 
import matplotlib.pyplot as plt

img = cv2.imread('../images/lena-computer-vision.png', cv2.IMREAD_GRAYSCALE)

numLinhas = img.shape[0]
numColunas = img.shape[1]
numOcorrenciaCor = [0]*256

for l in range(numLinhas):
    for c in range(numColunas):
        cor = img[l][c]
        numOcorrenciaCor[cor] += 1

escalaDeCinza = list(range(0,256))

plt.figure(1)
plt.stem(numOcorrenciaCor)
plt.title("Calculo de histograma")
plt.show
```

```python
plt.figure(2)
plt.stem(numOcorrenciaCor/(numLinhas*numColunas))
plt.title("Calculo de histograma")
plt.show
```

```python
limiar = 220

numLinhas = img.shape[0]
numColunas = img.shape[1]

for l in range(numLinhas):
    for c in range(numColunas):
        if img[l][c] > limiar:
            img[l][c] = 255
        else: img[l][c] = 0
```
